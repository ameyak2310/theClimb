```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
plt.figure()
```

```python
t  = np.linspace(0, 2*np.pi, 50)
x = np.sin(t)
y = np.cos(t)
```

```python
plt.plot(x)
plt.title("Sine")
```

```python
plt.figure()
plt.plot(y)
plt.title("Cosine")
```

```python
x = np.linspace(0,2*np.pi, 50)
y1 = np.sin(x)
y2 = np.sin(2*x)
```

```python
plt.figure()
plt.plot(y1)
plt.plot(x, y1, 'r-o')
plt.plot(x, y1, 'ro')
plt.plot(x, y1, 'r+')
plt.clf()
```

```python
plt.plot(x, y1, 'g-o', x, y2, 'b-+')
plt.legend(['sin(x)','sin(2x)'])
```

```python
# Subplots
t = np.linspace(0,2*np.pi, 50)
x = np.sin(t)
y = np.cos(t)
plt.figure(figsize=(8,4))

plt.subplot(1,2,1)
plt.plot(x)

plt.subplot(1,2,2)
plt.plot(y)
```

```python
from numpy.random import randint
data = randint(10000, size=(10, 1000))
x = np.sum(data, axis=0)

plt.subplot(2,1,1)
plt.hist(x, color='r', density=True)

plt.subplot(2,1,2)
plt.hist(x, cumulative=True, color='b')
```
