```python
pip install ipythonblocks
from ipythonblocks import BlockGrid, colors, show_color, show_color_triple

#Creats Grid
grid = BlockGrid(5,4, fill=colors.MistyRose)

# Acess a cel from gird
Block = grid[0,0]

#Displays mentioned colors
show_color_triple(colors.Black)

# Gives RBG values
colors.AliceBlue

# Shows color from RGB values
show_color_triple((250,100,200))

# Change color of the block
Block.rgb = colors.DarkKhaki
# or
grid[0,1].rgb = colors.RosyBrown
grid[3,3] = colors.BlanchedAlmond
grid[-1,-1] = colors.BurlyWood

Block
grid

# For Loop
for block in grid.animate():
    block.rgb = colors.Black 

# For loop w/ condition  
for block in grid.animate():
    if block.row == 2:
        block.rgb = colors.PaleTurquoise 

# For loop w/ multiple conditions
for block in grid.animate():
    if block.row == 0:
        block.rgb = colors.BlueViolet
    elif block.row == 2:
        block.rgb = colors.Azure
    else:
        block.rgb = colors.PaleTurquoise

# Range function
for row_index in range(grid.height):
    grid[row_index,1].rgb = colors.NavajoWhite

# or statement
for block in grid.animate():
    if block.col == 0 or block.col == 2:
        block.rgb = colors.Red

# and statement
for block in grid.animate():
    if block.row == 0 and block.rgb == colors.Brown:
        block.rgb = colors.Red

# animate with range
for index_row in range(grid.width):
    for index_col in range(grid.height):
        grid[index_col,index_row] = colors.Ivory
        grid.flash()

grid = BlockGrid(20,20)
for index_row in range(grid.height // 2, grid.height):
    for index_col in range( grid.width // 2, grid.width):
        grid[index_row,index_col].rgb = colors.Aquamarine
        grid.flash()

# Slicing
grid[8:12,7:13] = colors.Chartreuse        
```
